```python
# Import
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Parameters
n = 30
num = 99991
num_threads = 12
max_number = 900000
step = 70001

experiment = 'Experiment_1'
optimization = 'Od'
csv_name = "output_inefficient"

csv_path = "./Data/" + experiment + "/" + csv_name + "_" + optimization + ".csv"
```

```python
CSV = pd.read_csv(csv_path)
```

```python
# Execution time
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20, 6))

y_V = []

ci = []

for number in range(step, max_number+1, step):

    y_V.append(CSV[CSV["number"] == number]["execution_time"].mean())
    ci.append(1.96 * (CSV[CSV["number"] == number]["execution_time"].std() / math.sqrt(n)))


plt.errorbar(x=range(step, max_number+1, step), y=y_V, fmt='o', yerr=ci, markersize=3, capsize=4, color='b', label="95% CI")

sns.lineplot(x=range(step, max_number+1, step), y=y_V, marker='o', label="Mean")
    
plt.xlabel('Number used')
plt.ylabel('Mean Execution time in milliseconds [ms]')

plt.legend()
plt.show()
```
